On va dans les prochaines cellules chercher à faire un peu plus de sens avec des données qui n'en ont pas tel quel**

```python
#Au lieu de simplement parler du nombre total de pièces dans un quartier
#On va plutot parler du nombre de pièces par maison du quartier
data["rooms_in_house"] = data["total_rooms"] / data["households"]
data["rooms_in_house"]
```

```python
#Au lieu de simplement parler du nombre total de chambres dans un quartier
#On va plutot parler du nombre de chambres par rapport au nombre de pièces
data["bedrooms_in_rooms"] =  data["total_bedrooms"] / data["total_rooms"]
data["bedrooms_in_rooms"]
```

```python
#Au lieu de simplement parler du nombre total de chambres dans un quartier
#On va plutot parler du nombre de chambres par maison du quartier
data["bedrooms_in_house"] = data["total_bedrooms"] / data["households"]
data["bedrooms_in_house"]
```

```python
#On va regarder si le nombre de personne dans une maison a une corrélation avec son prix
data["population_per_household"] = data["population"] / data["households"]
data["population_per_household"]
```

```python
#Et maintenant on vérifie les corrélations avec les nouvelles valeurs trouvées
correlation = data.corr()
correlation["median_house_value"]
```

```python
correlation = data.corr()
plt.figure(figsize=(30,20))
plt.subplot(111)
sns.heatmap(correlation, annot=True, linewidths=1.0)
plt.show()
```

**On voit plusieurs corrélations apparaître, en plus de celles que l'on a déjà explicité :
La latitude en négatif : c'est à dire que plus on va vers le nord, plus le prix diminue, dans cet exemple c'est logique car cela signifie que l'on s'éloigne de l'eau, et on sait que la proximité à l'océan joue
bedrooms_in_rooms en négatif : cela signifie que plus l'on a de pièces qui ne sont pas des chambres, plus la valeur augmente
Et rooms_in_house : qui signifie que plus on a de pièces dans une maison plus elle a de valeurs**
