```python
import matplotlib.pyplot as plt
import pandas as pd


nom_colonnes = ['consommation', 'cylindres', 'deplacement',
                'puissance', 'poids', 'acceleration', 'annee', 'origine', 'nom']

data = pd.read_csv('Jeux_de_donnees\\auto-mpg.csv', na_values='?',
                     index_col=None, header=None, skiprows=1, names=nom_colonnes)

# On suprime les ligne avec des valeurs manquantes.
data = data.dropna()


X = data[['consommation', 'cylindres', 'deplacement',
                'puissance', 'poids', 'acceleration', 'annee']]
nom = data['nom'].values

X
```

```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html#scipy.cluster.hierarchy.linkage
from scipy.cluster import hierarchy
Z = hierarchy.linkage(X, 'ward')
```

```python
plt.figure(figsize=(20, 7))
dn = hierarchy.dendrogram(Z, labels = list(nom))
plt.xticks(rotation=90);
```

```python
data_head = data.head(10)
X = data_head[['consommation', 'cylindres', 'deplacement',
                'puissance', 'poids', 'acceleration', 'annee']]

nom = data_head['nom'].values

Z = hierarchy.linkage(X, 'ward')
plt.figure(figsize=(20, 7))
dn = hierarchy.dendrogram(Z, labels = list(nom))
plt.xticks(rotation=45);
```

```python
cutree = hierarchy.cut_tree(Z, height=5000)
```
