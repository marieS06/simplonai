```python
# Exploration des repas de thanksgiving
# â€¢Utiliser la mÃ©thode Series.value_counts() pour afficher combien de 
# fois chaque rÃ©sultat apparaÃ®t pour la question et colonne 'Whatistypicallythe main dishat yourThanksgiving dinner?'


data = pd.read_csv("thanksgiving.csv", encoding ='latin1')

pa=data['What is typically the main dish at your Thanksgiving dinner?'].value_counts()
print(pa)
```

```python
# â€¢Afficher la colonne 'Do you typically have gravy?' pour les
# lignes du dataframe data pour lesquelles la colonne 'What ist ypically the main dish at yourThanksgiving dinner?'
# vaut 'Tofurkey' pour la dinde de tofu.
# â€¢CrÃ©er un filtre qui sÃ©lectionne seulement les lignes du dataframe data oÃ¹ 
# 'What ist ypically the main dishat your Thanksgiving dinner?' vaut 'Tofurkey'
# â€¢Assigner le rÃ©sultat Ã  la variable
# tofurkey
# â€¢Afficher la colonne 'Do you typically have gravy?' de ce nouveau dataframe tofurkey

tofurkey=data[data['What is typically the main dish at your Thanksgiving dinner?']=="Tofurkey"]

c=tofurkey['Do you typically have gravy?']

print(c)
```

```python
# Tendances des desserts pour Thanksgiving
# â€¢CrÃ©er un objet Series indiquant avec des boolÃ©ens les valeurs de la colonne 
# 'Whichtype of pie is typicallyserved at your Thanksgiving dinner? Please select all that apply. -Apple' qui sont nulles.
# Assigner le rÃ©sultat Ã  la variable apple_isnull.â€¢CrÃ©er un objet Seriesindiquant avec des boolÃ©ens les valeurs
# de la colonne 'Whichtype of pie istypicallyservedat yourThanksgiving dinner? Pleaseselect all thatapply. -Pumpkin'
# qui sont nulles. Assigner le rÃ©sultat Ã  la variable pumpkin_isnull.â€¢CrÃ©er un objet Seriesindiquant avec des 
# boolÃ©ens les valeurs de la colonne 'Whichtype of pie istypicallyservedat yourThanksgiving dinner? Pleaseselect all
# thatapply. -Pecan' qui sont nulles. Assigner le rÃ©sultat Ã  la variable pecan_isnull.
# â€¢Combiner ces 3 objets Seriesavec l'opÃ©rateur & et assigner le rÃ©sultat Ã  la variable pies.
# â€¢Afficher les valeurs uniques et combien de fois elles apparaissent dans la colonne de pies

# s1 = data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple']
apple_isnull = pd.Series( data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple'] )
apple_isnull = apple_isnull.isnull() 

pumpkin_isnull = pd.Series( data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin'] )
pumpkin_isnull = pumpkin_isnull.isnull() 

pecan_isnull = pd.Series( data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan'] )
pecan_isnull = pecan_isnull.isnull() 

pies=(apple_isnull & pumpkin_isnull & pecan_isnull)
pies.unique()

print( len(pies[pies==True]) )
```
