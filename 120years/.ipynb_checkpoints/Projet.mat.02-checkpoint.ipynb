```python
import pandas as pd
reviews=pd.read_csv( 'fandango_scores.csv' )
reviews
```

```python
colonne=['FILM','RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
norm_reviews=reviews[colonne].head()
print(norm_reviews)
```

```python
import matplotlib.pyplot as plt
from numpy import arange


col = ['RT_user_norm', 'Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
bar_heights = norm_reviews[col].iloc[0].values

bar_positions = arange(5) + 1
fig, ax= plt.subplots()
ax.bar(bar_positions, bar_heights, 0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt
from numpy import arange

fig, ax= plt.subplots()
ax.bar(bar_positions, bar_heights, width= 0.5)
col = ['RT_user_norm', 'Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
bar_heights = norm_reviews[col].iloc[0].values

bar_positions = arange(5) + 1
tick_positions = range(1,6)

ax.set_xticks(tick_positions)
ax.set_xticklabels(col,rotation=90)
plt.title('Moyenne des notes utilisateurs pour le film Avengers: Age of Ultron(2015)')    
plt.ylabel('Note moyenne')
plt.xlabel('Sources de notation')


plt.show()
```

```python
import matplotlib.pyplot as plt
from numpy import arange

col = ['RT_user_norm', 'Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
num_cols = norm_reviews[col].iloc[0].values
bar_positions = arange(5) + 1
tick_positions = range(1,6)
fig, ax= plt.subplots()
#ax.barh(bar_positions, 0.5)
ax.barh(bar_positions, num_cols,0.5)

ax.set_yticks(tick_positions)
ax.set_yticklabels(col)
ax.set_title('Note moyenne utilisateurs pour le film Avengers: Age of Ultron(2015)')
ax.set_xlabel( 'Note moyenne')
ax.set_ylabel('Sources de notes')
num_cols = ['RT_user_norm', 'Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']


plt.show()

```

```python
#Intervertir les axes

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax1.scatter(reviews["Fandango_Ratingvalue"], reviews["RT_user_norm"])
ax1.set_xlabel( 'Fandango' )
ax1.set_ylabel( 'RottenTomatoes')
ax2 = fig.add_subplot(2,1,2)
ax2.scatter(reviews["RT_user_norm"], reviews['Fandango_Ratingvalue' ])
ax2.set_xlabel('RottenTomatoes')
ax2.set_ylabel('fandango')

plt.show()
```

```python
#Comparaison de corrÃ©lations

fig = plt.figure()
ax1 = fig.add_subplot(3,1,1)
ax1.scatter(reviews["Fandango_Ratingvalue"], reviews["RT_user_norm"])
ax1.set_xlabel( 'Fandango' )
ax1.set_ylabel( 'RottenTomatoes')
ax1.set_xlim(0,5)
ax1.set_ylim(0,5)

ax2 = fig.add_subplot(3,1,2)
ax2.scatter(reviews["RT_user_norm"], reviews['Fandango_Ratingvalue' ])
ax2.set_xlabel( 'RottenTomatoes' )
ax2.set_ylabel(  'fandango')
ax2.set_xlim(0,5)
ax2.set_ylim(0,5)

ax3 = fig.add_subplot(3,1,3)
ax3.scatter(reviews["RT_user_norm"], reviews['Fandango_Ratingvalue' ])
ax3.set_xlabel( 'RottenTomatoes' )
ax3.set_ylabel(  'fandango')
ax3.set_xlim(0,5)
ax3.set_ylim(0,5)
```
