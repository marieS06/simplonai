```python
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import arange

sns.set_style("darkgrid")
# dessin en carré
col = ['RT_user_norm', 'Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
bar_heights = norm_reviews[col].iloc[0].values

bar_positions = arange(5) + 1
fig, ax= plt.subplots()
ax.bar(bar_positions, bar_heights, 0.5)

plt.show()
# Diagrammes en barre
```
