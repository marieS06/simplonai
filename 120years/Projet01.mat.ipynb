```python
step = 12
mycolor = ['r','b','g','C1','k']
unrate['DATE']=pd.to_datetime(unrate['DATE'])
unrate['MONTH']=unrate['DATE'].dt.month
plt.figure(figsize=(10, 6))
plt.xticks(rotation=90)

for n in range(0,5):
    myannee = 1948+n
    plt.plot(unrate["MONTH"][0+n*step:(1+n)*step],unrate["UNRATE"][0+n*step:(1+n)*step]
             ,mycolor[n],label=myannee,markersize=1.8)

    
plt.title('Evolution du taux de chômage de 1948 à 1952')    
plt.ylabel('Taux de chômage (en %)')
plt.xlabel('Mois (de 1 à 12)')
plt.legend(loc='upper left')

# à ne pas confondre avec annee
#        [0+n*step:(1+n)*step] avec n allant de 0 à 5 et step = 12
# n = 0 , premier tour -> [0+0*step:(1+0)*step] -> [0:step] -> [0:12]
# n = 1 , deuxieme tour -> [0+1*step:(1+1)*step] -> [1*step:2*step] -> [12:24]
# n = 2 , troisieme tour -> [0+2*step:(2+1)*step] -> [2*step:3*step] -> [24:36]
# etc....
```

```python
MONTH = pd.to_datetime(unrate['DATE']).dt.month

step = 12
mycolor = ['r','b','g','C1','k']
monthname = ['Jan','Fev','Mar','Avr','Mai','Jui','Jlt','Aou','Sep','Oct','Nov','Dec']

plt.figure(figsize=(10, 6))
plt.xticks(rotation=90)


for n in range(0,5):
    myannee = 1948+n
    plt.plot(MONTH[0+n*step:(1+n)*step],unrate["UNRATE"][0+n*step:(1+n)*step],mycolor[n],label=myannee,markersize=1.8)
    

    
plt.title('Evolution du taux de chômage de 1948 à 1952')    
plt.ylabel('Taux de chômage (en %)')
plt.xlabel('Mois')
plt.xticks(np.arange(1,13),monthname)
plt.grid()

plt.xlim(1,12)
plt.ylim(0,9)

plt.legend(loc='upper left')
# plt.legend(loc=2)

#        [0+n*step:(1+n)*step] avec n allant de 0 à 5 et step = 12
# n = 0 , premier tour -> [0+0*step:(1+0)*step] -> [0:step] -> [0:12]
# n = 1 , deuxieme tour -> [0+1*step:(1+1)*step] -> [1*step:2*step] -> [12:24]
# n = 2 , troisieme tour -> [0+2*step:(2+1)*step] -> [2*step:3*step] -> [24:36]
# etc....
```
