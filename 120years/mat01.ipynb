```python
import pandas as pd
donnees=pd.read_csv("120-years-of-olympic-history-athletes-and-results.csv", index_col=[1])
donnees.head()
```

```python
donnees.isnull()
```

```python
donnees.isnull().sum()
```

```python
donnees['Medal']=donnees['Medal'].fillna("None")
donnees
```

```python
donnees_NaN_supprimes = donnees.dropna()
donnees_NaN_supprimes = donnees_NaN_supprimes.drop_duplicates()
donnees_NaN_supprimes.isnull().sum()
```

```python
moyennes = donnees[['Age','Height','Weight']].mean()
donnees[['Age','Height','Weight']] = donnees[['Age','Height','Weight']].fillna(moyennes)
donnees_NaN_remplaces=donnees.drop_duplicates()
print(donnees_NaN_remplaces.isnull().sum())
print()
print(moyennes) # celui la est remplacer par la  moyenne
```

```python
import matplotlib.pyplot as plt
import numpy as np
# xy : les coordonnées du point sur lequel la flèche doit pointer, 
# xytext : les coordonnées où placer le texte grâce à l’option 
# arrowprops : les propriétés de la flèche permettant de relier votre point d’intérêt au texte 
# L’option arrowprops est optionnelle et contient un dictionnaire spécifiant les propriétés de la flèche à afficher.
# Ici, on ne donne qu’une option, on veut que la flèche soit noire, avec facecolor.


plt.figure()
plt.subplot(111)

plt.plot(donnees_NaN_supprimes["Height"],donnees_NaN_supprimes["Weight"],
        marker='*',color="black", linestyle='',markersize=2.5, label="Poids_vs_Taille")

plt.plot(donnees_NaN_supprimes["Height"],donnees_NaN_supprimes["Age"],
        marker='v',color="blue", linestyle='',markersize=2.5, label="Age_vs_Taille")

plt.plot(donnees_NaN_supprimes["Height"], age_moyen, linestyle='--',color='red', label="Moyenne_age")


plt.plot(donnees_NaN_supprimes["Height"],[np.mean(donnees_NaN_supprimes["Age"])]*len(donnees_NaN_supprimes["Height"]))

plt.text(180,200,"y=Poids, x=Taille",size=13,bbox=dict(boxstyle="round", fc="white",ec="r"))
         
plt.title("Athletes des jos")
plt.ylabel("Poid/Age")
plt.xlabel("Taille")

plt.legend(loc='upper left')# placer en haut à gauche pas oublier label
plt.show()
```

```python
sports_10=donnees_NaN_remplaces['Sport'].value_counts()[:10].keys()
donnees_bar=donnees_NaN_remplaces.loc[donnees_NaN_remplaces['Sport'].isin(sports_10)].groupby(['Sport','Sex']).size().unstack(1)
donnees_bar
```

```python
import matplotlib.pyplot as plt

pos=list(range(len(donnees_bar['F'])))
width = 0.25
plt.figure()
plt.subplot(111)
plt.bar(pos, donnees_bar['F'], width, color='red',label="Femmes")
plt.bar([p + width for p in pos], donnees_bar['M'], width, color='blue', label="Hommes")
plt.xticks([p + 0.125 for p in pos], donnees_bar.index,rotation = 90)
plt.ylabel("Nombre d'athlètes")
plt.legend()
plt.title("Nombre d'athlètes hommes et femmes dans 10 disciplines")
plt.show
```

```python
valeurs_hommes=donnees_bar["M"]
valeur_femmes=donnees_bar['F']
width=0.35
plt.figure()
plt.subplot(111)
plt.bar(donnees_bar.index,valeurs_hommes, width, label='Hommes')
plt.bar(donnees_bar.index, valeur_femmes, width, bottom=valeurs_hommes, label='Femmes')
plt.ylabel('Nombre')
plt.title(' Nombre d athlètes par discipline en fonction du genre')
plt.xticks(donnees_bar.index, rotation=90)
plt.legend()
plt.show()
```
