# Detection de visages

```python
import cv2 as cv
```

#### On charge les classificateurs en cascade pré-entrainés

```python
face_cascade = cv.CascadeClassifier(
    'Detection de visages\haarcascade_frontalface_default.xml')
```

#### On charge l'image
On va la covertir en niveaux de gris pour la passer à notre classificateur en cascade

```python
img = cv.imread('Detection de visages\dan.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.subplot(111)
plt.imshow(img)
plt.title("Photo de Dan")
plt.axis('off');
```

#### Exécution de la détection de visage avec pour :
- paramètre d'échelle 1.1
- nombre de voisins 8

```python
# detectMultiScale(image, scale factor, number of neighbors)
# detectMultiScale(image, paramètre d'échelle, nombre de voisins)
faces = face_cascade.detectMultiScale(gray, 1.1, 8)
```

On obtient une liste de coordonnées nous donnant :
- les coordonnées du coin superieur gauche du visage détecté (x et y)
- ainsi que sa largeur et sa hauteur.

Attention des faux postifs peuvent apparaitre. On les identifira notament grace à leur petite largeur et hauteur.

```python
faces
```

#### Pour une photo à 2 visages j'obtiens 2 listes de coordonnées

```python
img = cv.imread('Detection de visages\\bradangelina.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.figure(figsize=(10,5))
plt.subplot(111)
plt.imshow(img)
plt.title("Photo de Dan")
plt.axis('off');

faces = face_cascade.detectMultiScale(gray, 1.1, 8)
```

```python
faces
```

#### On affiche les visages détèctés.

```python
# affichage des visages
i = 0
for face in faces:
    x, y, w, h = face
    
    # dessiner le rectangle sur l'image principale
    cv.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    
# affiche l'image principale
cv.imshow('image principale', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

#### Maintenant on va découper les visages

```python
# affichage des visages
i = 0
for face in faces:
    x, y, w, h = face
    
    # dessiner le rectangle sur l'image principale
    # (0, 255, 0), 2) permet de choisir la couleur du rectangle
    cv.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    
    # extraire les visages de l'image principale
    # OpenCV et Numpy: y <-> row et x <-> col
    face = img[y:y+h, x:x+w]
    
    # afficher face0, face1, face2, etc...
    cv.imshow('face{}'.format(i), face)
    cv.waitKey(0)
    cv.destroyAllWindows()
    i += 1
```
