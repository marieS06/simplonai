```python
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv("Jeux_de_donnees/iris.csv",
                   delimiter=";", encoding="latin-1")
iris
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=10)
X=iris[iris.columns[:-1]]

clustering = dbscan.fit_predict(X)
y_pred = dbscan.fit_predict(X)
```

Les clusters d'appatenance de chacune de mes observation. Le -1 correspond aux observations n'appartenant à aucun cluster.

```python
iris['predict']=dbscan.labels_
iris['core'] = 'no_core'
iris.iloc[dbscan.core_sample_indices_, -1]='core'
iris
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=[20,20])
plt.subplot(221)
sns.scatterplot(data=iris, y='Longueur des sépales',
                x='Largeur des sépales',
                hue='predict', style="core", palette="Set1")
plt.title("Longueur des sépales en focntion de la largeur et coloré par group identifier")

plt.subplot(222)
sns.scatterplot(data=iris, y='Longueur des sépales',
                x='Largeur des sépales',
                hue='Espèce', style="core", palette="Set1")
plt.title("Longueur des sépales en focntion de la largeur et coloré par espèce")


plt.subplot(223)
sns.scatterplot(data=iris, y='Longueur des pétales',
                x='Largeur des pétales',
                hue='predict', style="core", palette="Set1")
plt.title("Longueur des pétales en focntion de la largeur et coloré par group identifier")

plt.subplot(224)
sns.scatterplot(data=iris, y='Longueur des pétales',
                x='Largeur des pétales',
                hue='Espèce', style="core", palette="Set1")
plt.title("Longueur des pétales en focntion de la largeur et coloré par espèce")

plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(X, clustering.labels_)
```

```python
min_samples = [i for i in range(1,15)]
silhouette_min_sample = []

iris = pd.read_csv("Jeux_de_donnees/iris.csv", delimiter=";", encoding="latin-1")
X=iris[iris.columns[:-1]]

for i in min_samples:
    
    clustering = DBSCAN(eps=0.5, min_samples=i)
    clustering.fit(X)
    
    silhouette_min_sample.append(silhouette_score(X, clustering.labels_))

silhouette_min_sample
```

```python
data = pd.DataFrame({'min_samples':min_samples, 'silhouette_min_sample':silhouette_min_sample})
sns.lineplot(data=data, x="min_samples", y="silhouette_min_sample", marker='o');
```

```python
epsilon = np.arange (0.4, 1.8, 0.2)
#epsilon = [0.4, 0.6, 0.8, 1, 1.5, ]
silhouette_epsilon = []

iris = pd.read_csv("Jeux_de_donnees/iris.csv", delimiter=";", encoding="latin-1")
X=iris[iris.columns[:-1]]

for i in epsilon:
    
    clustering = DBSCAN(eps=i, min_samples=5)
    clustering.fit(X)
    
    silhouette_epsilon.append(silhouette_score(X, clustering.labels_))

silhouette_epsilon
```

```python
data = pd.DataFrame({'epsilon':epsilon, 'silhouette_epsilon':silhouette_epsilon})
sns.lineplot(data=data, x="epsilon", y="silhouette_epsilon", marker='o');
```
