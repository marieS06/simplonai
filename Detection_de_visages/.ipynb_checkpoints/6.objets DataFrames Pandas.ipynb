## Utiliser des index personnalisés

- Utiliser la méthode set_index pour indéxer notre DataFrame fandango avec la colonne 'FILM'. A préciser que nous souhaitons conserver la colonne 'FILM'. De plus nous souhaitons garder le DataFrame d'origine et en créer un nouveau que nous assignerons à la variable fandango_films.
- Afficher l'index du DataFrame fandango_films à l'iade de l'attribut index.

```python
fandango_films = fandango.set_index('FILM', drop=False)
fandango_films
```

## Sélectionner des valeurs depuis un index

- Sélectionner les films suivants depuis le DataFrame fandango_films (dans l'ordre suivant) et assigner les à la variable some_movies:
>- "Top Five (2014)"
>- "Black Sea (2015"
>- "Taken 3 (2015)"
- Afficher le résultat.

```python
movies = ["Top Five (2014)","Black Sea (2015)","Taken 3 (2015)"]
some_movies = fandango_films.loc[movies]
some_movies
```

## Méthode Apply() sur les colonnes d'un DataFrame

- Obtenir le dataframe float_df ne contenant que les colonnes décimales float64 à l'aide de l'attribut .dtypes
- Utiliser la méthode apply() sur ce dataframe float_df afin d'obtenir l'écart type chaque colonne et assigner le résultat à la variable exo_df.
- Afficher les premières lignes du résultat.

```python
fandango_films.dtypes
```

```python
types = fandango_films.dtypes
float_columns = types[types.values == 'float64'].index
float_columns
```

```python
float_df = fandango_films[float_columns]
float_df.head()
```

```python
import numpy as np
exo_df = float_df.apply(lambda col: np.std(col))
exo_df.head()
```

```python
type(exo_df)
```

## Méthode Apply() sur les lignes d'un DataFrame

- Utiliser la méthode apply() pour calculer la moyenne (méthode mean() de la librairie Numpy) des valeurs de chaque film pour 'RT_user_norm' et 'metacritic_user_nom' puis assigner le résultat à la variable rt_mt_means.
- Afficher les 5 premières valeurs de rt_mt_means.

```python
import numpy as np

rt_mt_user = float_df[['RT_user_norm', 'Metacritic_user_nom']]
rt_mt_means = rt_mt_user.apply(lambda row: np.mean(row), axis=1)
```

```python
print(rt_mt_means[0:5])
```
