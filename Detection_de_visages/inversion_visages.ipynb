```python
import cv2 as cv
import numpy as np

face_cascade = cv.CascadeClassifier("donnees/Detection_de_visages/haarcascade_frontalface_default.xml")
img = cv.imread('donnees/Detection_de_visages/obama.jpg')
gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.subplot(111)
plt.imshow(gray,cmap="gray")
plt.title("Photo du couple Obama")
plt.axis('off');
```

```python
faces = face_cascade.detectMultiScale(gray, 1.1, 8)
if faces.size/4 != 2 :
    print("L'image ne contient pas exactement 2 visages")
else :
    # echange des deux visages
    print(faces)

    x1,y1,w1,h1 = faces[0]
    x2,y2,w2,h2 = faces[1]

    face1 = img[y1:y1+h1, x1:x1+w1]
    face2 = img[y2:y2+h2, x2:x2+w2]

    face1 = cv.resize(face1, (w2,h2), interpolation= cv.INTER_LINEAR)
    face2 = cv.resize(face2, (w1,h1), interpolation= cv.INTER_LINEAR)

    img[y1:y1+h1, x1:x1+w1] = face2
    img[y2:y2+h2, x2:x2+w2] = face1
    
    # affiche l'image principale
    cv.imshow('image principale', img)
    cv.waitKey(0)
    cv.destroyAllWindows()
```
