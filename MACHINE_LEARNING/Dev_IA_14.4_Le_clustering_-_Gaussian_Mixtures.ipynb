Nous ne pouvons pas utiliser l'inertie ou le score de silhouette car ils supposent tous deux que les clusters sont sphériques. Au lieu de cela, nous pouvons essayer de trouver le modèle qui minimise un critère d'information théorique tel que le critère d'information bayésien (BIC) ou le critère d'information d'Akaike (AIC) : 

${BIC} = {\log(m)p - 2\log({\hat L})}$

${AIC} = 2p - 2\log(\hat L)$

* $m$ est le nombre d'instances.
* $p$ est le nombre de paramètres appris par le modèle.
* $\hat L$ est la valeur maximisée de la fonction de vraisemblance du modèle. C'est la probabilité conditionnelle des données observées $\mathbf{X}$, étant donné le modèle et ses paramètres optimisés. 

BIC et AIC pénalisent tous deux les modèles qui ont plus de paramètres à apprendre (par exemple, plus de clusters) et récompensent les modèles qui correspondent bien aux données (c'est-à-dire les modèles qui donnent une forte probabilité aux données observées).

```python
print('BIC : ', gm.bic(X), ', AIC : ', gm.aic(X))
```

#### On test différents modèles cette fois ci en mettant des contraintes sur la matrice variance-covariance.

```python
gm_diag = GaussianMixture(n_components=3, n_init=10,
                          covariance_type="diag", random_state=42)
gm_tied = GaussianMixture(n_components=3, n_init=10,
                          covariance_type="tied", random_state=42)
gm_spherical = GaussianMixture(n_components=3,
                               n_init=10, covariance_type="spherical", random_state=42)
gm_diag.fit(X)
gm_tied.fit(X)
gm_spherical.fit(X)
print('diag BIC: ', gm_diag.bic(X), ', diag AIC: ', gm_diag.aic(X))
print('diag BIC: ', gm_tied.bic(X), ', diag AIC: ', gm_tied.aic(X))
print('diag BIC: ', gm_spherical.bic(X), ', diag AIC: ', gm_spherical.aic(X))
```
