```python
# Tendances des desserts pour Thanksgiving
# â€¢CrÃ©er un objet Series indiquant avec des boolÃ©ens les valeurs de la colonne 
# 'Whichtype of pie is typicallyserved at your Thanksgiving dinner? Please select all that apply. -Apple' qui sont nulles.
# Assigner le rÃ©sultat Ã  la variable apple_isnull.â€¢CrÃ©er un objet Seriesindiquant avec des boolÃ©ens les valeurs
# de la colonne 'Whichtype of pie istypicallyservedat yourThanksgiving dinner? Pleaseselect all thatapply. -Pumpkin'
# qui sont nulles. Assigner le rÃ©sultat Ã  la variable pumpkin_isnull.â€¢CrÃ©er un objet Seriesindiquant avec des 
# boolÃ©ens les valeurs de la colonne 'Whichtype of pie istypicallyservedat yourThanksgiving dinner? Pleaseselect all
# thatapply. -Pecan' qui sont nulles. Assigner le rÃ©sultat Ã  la variable pecan_isnull.
# â€¢Combiner ces 3 objets Seriesavec l'opÃ©rateur & et assigner le rÃ©sultat Ã  la variable pies.
# â€¢Afficher les valeurs uniques et combien de fois elles apparaissent dans la colonne de pies

cl = 'Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - '
v1 = 'Apple'
v2 = 'Pumpkin'
v3 = 'Pecan'

apple_isnull = pd.isnull(data[cl + v1])
pumpkin_isnull = pd.isnull(data[cl + v2])
pecan_isnull = pd.isnull(data[cl + v3])


pies = apple_isnull & pumpkin_isnull & pecan_isnull

# On compte les Falses a cause du isnull c.a.d que l'on compte seulement 
# ceux qui ne mange pas de tarte a Thanksgiving
# On suppose que personnes ne mangent les 3 types de tarte pour Thanksgiving
# c.a.d que apple_isnull & pumpkin_isnull & pecan_isnull n'est jamais Ã©gale Ã :
# apple_isnull & pumpkin_isnull & pecan_isnull  != True & True & True
# Les seules possibilitÃ©s sont : 
# True & False & False = False
# False & True & False = False
# False & False & True = False
# False & False & False = True
# Mais Jamais : True & True & True = True
# Sinon on compterais de la mama maniere ceux qui:
# n'ont pas mangÃ© de tarte 
# et ceux qui ont mangÃ© les 3 types de tartes

print(pies.unique())
pies.value_counts()
```

```python
# Convertir l'Ã¢ge en une valeur numÃ©rique
# â€¢Ecrire une fonction qui convertit une chaine de caractÃ¨res en une valeur entiÃ¨re. 
# Cela permettra de convertir les valeurs de la colonne 'Age' en entiers. 
# Cette fonction prendra en paramÃ¨tre une chaine de caractÃ¨res (les valeurs actuelles de la colonne 'Age')
# â€¢Utiliser la fonction isnull() pour vÃ©rifier si les valeurs sont nulles. 
# Ajouter une condition if qui retourne 'None' si la valeur est nulle
# â€¢SÃ©parer les chaines de caractÃ¨res en fonction de l'espace (' ') 
# et extraire le premier Ã©lÃ©ment de la liste rÃ©sultante (mÃ©thode split())
# â€¢Remplacer le caractÃ¨re '+' dans le rÃ©sultat avec une chaine de caractÃ¨res vide en remplacement pour 
# le supprimer (mÃ©thode replace())â€¢utiliser int() pour convertir le rÃ©sultat en entierâ€¢Retourner le rÃ©sultat
# â€¢Utiliser la mÃ©thode Series.apply() pour appliquer la fonction Ã  chaque valeur de la colonne 'Age' 
# du dataframe data.
# â€¢Assigner le rÃ©sultat Ã  la nouvelle colonne int_agedu dataframe data
# â€¢Appeler la mÃ©thode Series.
# describe() sur la colonne 'int_age' du dataframe data et afficher le rÃ©sultat
import pandas as pd
import numpy as np


def agees(dato):
    
    if pd.isnull(dato):
        return None
    dato = dato.split()[0]
    dato = dato.replace("+","")
    return int(dato)

data["int_age"] = data['Age'].apply(agees)
data["int_age"].describe()
        
```

```python
import pandas as pd
import numpy as np


def aei(ae):
    
    if pd.isnull(ae):
        return None
    ae = ae.split(" ")[0]
    if ae == 'Prefer':
        return None
    ae = ae.replace("$","")
    ae = ae.replace(",","")
    return int(ae)

data["int_income"] = data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(aei)
data["int_income"].describe()
        
```
