```python
import os
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.model_selection import train_test_split
```

```python
DATA_DIR = 'jeux_de_donnees/enron'
target_names = ['ham', 'spam']
# On crée une liste tout les noms des sous dossier dans le dossier eron
subfolders = ['enron%d' % i for i in range(1,7)]

#On crée une liste X qui contiendra tout les email
X = []
#On crée une liste y qui contiendra 1 si spam 0 sinon.
y = []
```

```python
for subfolder in subfolders:
    
    #On crée une liste de tout les fichier contenue le dossier spam de chaque sous-dossier
    spam_files = os.listdir(os.path.join(DATA_DIR, subfolder, 'spam'))
    for spam_file in spam_files:
        with open(os.path.join(DATA_DIR, subfolder, 'spam', spam_file), encoding='ascii', errors='ignore') as f:
            X.append(f.read())
            # On ajouter 1 à notre liste target pour chaque spam lu dans la liste data.
            y.append(1)
            
    ham_files = os.listdir(os.path.join(DATA_DIR, subfolder, 'ham'))
    for ham_file in ham_files:
        with open(os.path.join(DATA_DIR, subfolder, 'ham', ham_file), encoding='ascii', errors='ignore') as f:
            X.append(f.read())
            y.append(0)
```

```python
X_1, X_2, y_1, y_2 = train_test_split(X, y, test_size=0.10, random_state=42)
```

```python
count_vec = CountVectorizer(stop_words={'english'})
X_counts = count_vec.fit_transform(X_2)
print(X_counts)
```

```python
type(X_counts)
```

```python
X_counts = pd.DataFrame(X_counts.toarray(),
                        columns = list(count_vec.vocabulary_.keys()))
```

```python
X_counts
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
cluster = [i for i in range(2,10)]
inertie = []
silhouette = []


for i in cluster:
    k = i
    clustering = KMeans(n_clusters=k, random_state=42)
    clustering.fit(X_counts)
    
    inertie.append(clustering.inertia_)
    silhouette.append(silhouette_score(X_counts, clustering.labels_))
    
    print('clustering.inertia_ : ', clustering.inertia_,
         'silhouette_score : ', silhouette_score(X_counts, clustering.labels_))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.DataFrame({'cluster':cluster, 'inertie':inertie,
                     'silhouette':silhouette})

plt.figure(figsize=[10,5])

plt.subplot(121)
sns.lineplot(data=data, x="cluster", y="inertie", marker='o')
plt.subplot(122)
sns.lineplot(data=data, x="cluster", y="silhouette", marker='o');
```

#### On va donc conserver 6 clusters

```python
k = 6
clustering = KMeans(n_clusters=k, random_state=42)
clustering.fit(X_counts)
```

#### Je viens joindre les label à mon dataframe

```python
X_counts['cluster_']=clustering.predict(X_counts)
```

#### Je regarde quel sont les dix mot qui caractérise le plus chacun des clusters

```python
for i in range(1,7):
    print('Cluster ', i-1, ' : ',
          X_counts[X_counts.cluster_==i-1].mean().sort_values(ascending=False).head(10).index.to_list(), '\n')
```

#### Classification hiérarchique

```python
# Je commence par récupérer les centre de gravité dans un DataFrame
# et je les répettes pour chaques obs de départ

cluster_centre = pd.DataFrame(clustering.cluster_centers_[clustering.labels_],
                      columns=X_counts.columns[:-1])
```

```python
cluster_centre
```

```python
from scipy.cluster import hierarchy

Z = hierarchy.linkage(cluster_centre, 'ward')

plt.figure(figsize=(20, 7))
plt.xticks(rotation=90)
dn = hierarchy.dendrogram(Z, truncate_mode='level', p=6)
```

#### Je decide de regrouper les clusters 3 par 3  pour en faire un deux clusters

```python
cutree = hierarchy.cut_tree(Z, height=2000)
cutree
```

```python
X_counts['cluster2_']=cutree
```

```python
X_counts['cluster2_'].value_counts()
```

```python
for i in range(1,3):
    print('Cluster ', i-1, ' : ',
          X_counts[X_counts.cluster2_==i-1].mean().sort_values(ascending=False).head(20).index.to_list(), '\n')
```
